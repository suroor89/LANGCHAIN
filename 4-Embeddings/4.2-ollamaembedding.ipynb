#### Ollama
Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

Ollama can run models locally, offering lower latency and better privacy when compared to cloud-only embedding providers. It exposes easy APIs for embedding generation and model management, and can be integrated with LangChain workflows to create semantic search, document retrieval, and similarity search pipelines.

With Ollama you can:
- use local or self-hosted models for generating embeddings
- persist vector indexes from text, documents, or knowledge bases
- combine embeddings with prompts to improve question answering and retrieval
- reduce reliance on external cloud services while keeping data under your control

In this notebook, we use Ollama to generate embeddings that can be stored and queried for downstream search and RAG tasks.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=OllamaEmbeddings(model="gemma:2b") ## by default it uses llama2
```

```python
embeddings
```

```python
text_lists=embeddings.embed_documents(
    [
       "Alpha is the first letter of Greek alphabet",
       "Beta is the second letter of Greek alphabet", 
    ]
)
```

```python
len(text_lists[0])
```

```python
text_lists[1]
```

```python
embeddings.embed_query("What is the first letter of Greek alphabet?")
```

```python
### Other Embedding Models
### https://ollama.com/blog/embedding-models
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
text = "This is a test document."
vector = embeddings.embed_query(text)
vector
```
