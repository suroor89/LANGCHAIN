#### Sentence Transformers on Hugging Face
Hugging Face sentence-transformers is a Python framework for state-of-the-art sentence, text and image embeddings. It provides ready-to-use models for semantic search, clustering, paraphrase detection, and retrieval tasks.

One of the embedding models is used in the `HuggingFaceEmbeddings` class. We have also added an alias for `SentenceTransformerEmbeddings` for users who are more familiar with directly using that package.

Key strengths of Sentence Transformers:
- pretrained models optimized for sentence-level semantic similarity
- support for many languages and model sizes
- fast inference with PyTorch or TensorFlow backends
- easy transition from embeddings to downstream tasks like RAG and semantic search

In this notebook, we use Hugging Face embeddings to convert text into dense vector representations that can be indexed, searched, and compared for similarity.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
text = "This is a test document."
vector = embeddings.embed_query(text)
vector
```

```python
len(vector)
```

```python
doc_result = embeddings.embed_documents([text, "This is not a test document."])
doc_result[0]
```
