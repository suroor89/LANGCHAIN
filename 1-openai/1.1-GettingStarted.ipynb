#### Getting started With Langchain And Open AI

LangChain is a framework for developing applications powered by large language models (LLMs). It enables developers to build sophisticated AI applications by chaining together LLM capabilities with other tools and data sources. OpenAI's GPT models provide state-of-the-art language understanding and generation capabilities that integrate seamlessly with LangChain.

In this quickstart we'll see how to:

- Get setup with LangChain, LangSmith and LangServe
- Use the most basic and common components of LangChain: prompt templates, models, and output parsers.
- Build a simple application with LangChain
- Trace your application with LangSmith
- Serve your application with LangServe

**Prerequisites:**
- Python 3.8 or higher
- OpenAI API key (get one at https://platform.openai.com/api-keys)
- Basic familiarity with Python and LLMs

**What you'll learn:**
This tutorial covers the fundamental building blocks for creating LLM-powered applications. By the end, you'll understand how to prompt models, parse their responses, create reusable templates, debug applications with LangSmith, and deploy them using LangServe. These tools form the foundation for building production-ready AI systems.

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
## Langsmith Tracing
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]=os.getenv("LANGCHAIN_TRACING_V2")
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-4o")
print(llm)
```

```python
## Input and get response from LLM
response=llm.invoke("What is LangChain?")
print(response)
```

```python
## Chatprompt template
from langchain_core.prompts import ChatPromptTemplate
prompt=ChatPromptTemplate.from_messages([("system","You are a helpful assistant."),
                                         ("user","{input}")
                                         ])
prompt
```

```python
chain=prompt | llm
response=chain.invoke({"input":"What is LangChain?"})
print(response)
```

```python
type(response)
```

```python
#stroutput Parser
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
chain=prompt | llm | parser
response=chain.invoke({"input":"What is LangChain?"})
print(response)
```
